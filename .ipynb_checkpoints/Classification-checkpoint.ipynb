```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
HFPA = pd.read_csv('HFPA_class.csv')
HFPA.info()
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

```python
logit = LogisticRegression(random_state = 42, max_iter = 1000)
cv_scores = []
for i in range(len(X_train.columns)-1):
    sfs = SequentialFeatureSelector(logit, n_features_to_select=i+1)
    sfs.fit(X_train, y_train)
    cv = cross_val_score(estimator=logit,
                         X = X_train[X_train.columns[sfs.get_support()]], 
                         y = y_train,
                         scoring = 'accuracy', cv = 5).mean()
    cv_scores.append(cv)
    print(X_train.columns[sfs.get_support()])
    print(cv)
```
